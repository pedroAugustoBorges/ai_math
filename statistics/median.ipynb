```python
import pandas as pd
import numpy as np
import copy
```

```python
dataset = np.array([
    148, 150, 151, 149, 150, 152, 149, 151, 180
])
```

### Median Implementation (No Numpy)

```python
len_data = len(dataset)
```

```python
n = int((len_data / 2 if len_data % 2 == 0 else (len_data + 1) / 2) - 1)
n
```

```python
def sort_list(dt):
    new_array = dt.copy()
    n = len(new_array)

    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if new_array[j] < new_array[min_idx]:
                min_idx = j
        new_array[i], new_array[min_idx] = new_array[min_idx], new_array[i]

    return new_array
```

```python
print(sort_list(dataset))
```

```python
median = sort_list(dataset)[n]
```

```python
print(median)
```
