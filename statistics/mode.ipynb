### Mode Implementation (No Numpy)

```python
def count_values(dt):
    frequency = {}

    for x in dt:
        x = int(x)
        if int(x) in frequency:
            frequency[x] = frequency[x] + 1
        else:
            frequency[x] = 1

    return frequency
```

```python
frequency = count_values(dataset)
```
