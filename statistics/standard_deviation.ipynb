## Standard Deviation


$$\sigma = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}}$$


```python
import pandas
import numpy
import math
```

```python
a = [14, 2, 5]
b = [9, 5, 10]

a_n = len(a)
```

```python
def mean_calculate(l):

    mean = 0
    for i in l:
        mean+= i / len(l)

    return mean
```

```python
a_mean = mean_calculate(a)
```

```python
a_mean
```

```python
def deviation(xi, mean):
    return [(value - mean) for value in xi]
```

```python
a_deviation = deviation(xi= a, mean=a_mean)
```

```python
print(a_deviation)
```

```python
def pow_numbers(xi):
    return [(value ** 2) for value in xi]
```

```python
a_pow = pow_numbers(a_deviation)
```

```python
print(a_pow)
```

```python
def sum_numbers(xi):
    aux = 0
    for x in xi:
        aux += x
    return aux
```

```python
a_sum = sum_numbers(a_pow)
```

```python
print(a_sum)
```

```python
variance = a_sum / a_n
```

```python
def square(xi):
    return xi ** 0.5
```

```python
standard_deviation = square(variance)
```

```python
print(standard_deviation)
```
