## Standard Deviation


$$\sigma = \sqrt{\frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}}$$


```python
import pandas
import numpy
import math
```

```python
a = [4, 5, 6]
b = [1, 5, 9]
```

```python
def mean_calculate(l):

    mean = 0
    for i in l:
        mean+= i / len(l)

    return mean
```
