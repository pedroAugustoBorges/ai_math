## Implemention Covariance (No NumPy)

```python
a = [4, 5, 12, 7]
b = [1, 5, 3, 9]
```

```python
n = len(a) - 1
```

```python
def mean(xs):
    return sum(xs) / len(xs) 
```

```python
mean_a = mean(a)
mean_b = mean(b)
print(mean_b)
```

```python
def plus_array(mean, xs):
    print('mean ->', mean)
    print('array ->',xs)
    centered = []
    
    for x in xs:
        centered.append((x - mean))
        
    return centered
```

```python
centered_a = plus_array(mean_a, a)
centered_b = plus_array(mean_b, b)
```

```python
def print_centered(og, centered):
    print(f'array original: {og} -> after centered: {centered}' )
```

```python
print_centered(a,centered_a)
print_centered(b, centered_b)
```

```python
def multiply_chain(x, y):
    if len(x) != len(y):
        print('Erro: Input vectors must have the same dimensions')
        return
   
    length = len(x)
   
    sum_of_products = []
   
    print(f'{x} -> {y}')
   
    for i in range(0, length):
       print(f'{x[i]} * {y[i]} = {(x[i] * y[i])}')
       sum_of_products.append((x[i] * y[i]))
       
    
    print(f"sum_of_products: {sum_of_products}")
    return sum_of_products
```

```python
def sum_products(xi):
    aux = 0
    for x in xi:
        aux += x
    return aux
```

```python
sum_of_products = sum_products(multiply_chain(centered_a, centered_b))
```

```python
print(sum_of_products)
```

```python

covariance = sum_of_products / n
```

```python
print(covariance)
```

### Covariance Calculation
**Formula:**
$$cov(X, Y) = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}{n-1}$$

**Steps for $a=[4, 5, 12, 7]$ and $b=[1, 5, 3, 9]$:**
1. Means: $\bar{a}=7.0$, $\bar{b}=4.5$
2. Sum of cross-products: $(10.5) + (-1.0) + (-7.5) + (0) = 2.0$
3. Division ($n-1$): $2.0 / 3 \approx 0.6667$
