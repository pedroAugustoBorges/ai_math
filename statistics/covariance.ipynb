## Implemention Covariance (No NumPy)

```python
a = [4, 5, 12, 3]
b = [1, 5, 3, 9]
```

```python
def mean(xs):
    return sum(xs) / len(xs) 
```

```python
mean_a = mean(a)
mean_b = mean(b)
```

```python
def plus_array(mean, xs):
    print('mean ->', mean)
    print('array ->',xs)
    centered = []
    
    for x in xs:
        centered.append((x - mean))
        
    return centered
```

```python
centered_a = plus_array(mean_a, a)
centered_b = plus_array(mean_b, b)
```
