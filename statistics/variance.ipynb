#### Variance ($\sigma^2$)


$$\sigma^2 = \frac{\sum_{i=1}^{n} (x_i - \bar{x})^2}{n}$$

```python
import pandas as pd
import numpy as np
import math
```

## Variance Implementation (No Numpy)

```python
a = [4, 5, 6]
b = [1, 5, 9]
```

```python
a_np = np.array(a)
b_np = np.array(b)
```

```python
mean_a = a_np.mean()
mean_b = b_np.mean()
```

```python
mean_a
```

```python
def centering(l, mean):
    new_observations = []
    for x in l:
        new_observations.append(((x - mean ) ** 2))

    return new_observations
```

```python
centering_values_a = centering(a, mean_a)
centering_values_b = centering(b, mean_b)
```

```python
print('centering_values_a:' ,centering_values_a)
print('centering_values_b:' ,centering_values_b)
```

```python
def variance(center):
    return sum(center) / len(center)
```

```python
var_a = variance(centering_values_a)
var_b = variance(centering_values_b)
```

```python
print('var_a:' ,var_a)
print('var_b' ,var_b)

```

```python
var_b
```

```python
std_a = math.sqrt(var_a)
```

```python
std_b = math.sqrt(var_b)
```
