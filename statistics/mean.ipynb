#### Arithmetic Mean ($\bar{x}$)

$$\bar{x} = \frac{\sum_{i=1}^{n} x_i}{n}$$

```python
import pandas as pd
import numpy as np
```

```python
dataset = np.array([
    148, 150, 151, 149, 150, 152, 149, 151, 180
])
```

### Mean Implementation (No NumPy)

```python
def mean_calculate(l):

    mean = 0
    for i in dataset:
        mean+= i / len(l)

    return mean
```

```python
mean_calculate(dataset)
```
