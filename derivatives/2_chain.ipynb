## Step 1 - Number derivate from definition:

$\frac {(x + h)² - x²}{h}  $

```python
def f(x):
    return x ** 2

```

```python
h = 1e-5
```

```python
numerical_derivate = (f(x + h) - f(x)) / h
```

```python
numerical_derivate
```

## Step 2 - Reproduce the algebre in python

$(h +x)^{2} = x^{2} + 2xh + h^{2}   $

```python
x = 3
h = 1e-5

term1 = x**2
term2 = 2 * x * h
term3 = h**2

expanded = term1 + term2 + term3
original = x ** 2
```

```python
print("Expanded:", expanded)
print("Original:", original)
print("Difference:", expanded - original)
```

```python
result = (expanded - original) / h
print("After dividing by h:", result)
```

```python
algebra_version = (2 * h * x + h**2) / h
print(algebra_version)
```

### Simplify:

$ = 2x + h$ because h is simplified in $\frac{2xh + h^{2}} {h}$

```python
simplified = 2*x + h
(print(simplified))
```
