# Learning derivate from scratch

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# x = np.linspace(0, 10, 100)

x = np.arange(1, 11)
```

```python
x
```

## Define a function

```python
def f_linear(x):
    return 5 * x


def f_non_linear(x):
    return x**2
```

```python
y = f_linear(x)
w = f_non_linear(x)
```

```python
print(y)
print(w)
```

## Plot vizualize linear function

```python
plt.plot(x, y)
plt.axhline(0)
plt.axvline(0)

plt.title("f(x) = 5x")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.show()
```

```python
plt.plot(x, w)
plt.axhline(0)
plt.axvline(0)

plt.title("f(x) = x**2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)

plt.show()
```

## Compute and vizualize scope

```python
x1 = 2
x2 = 3

y1 = f_linear(x1)
y2 = f_linear(x2)

slope = (y2 - y1) / (x2 - x1)

print(f"(y2 - y1) = {y2 - y1} / (x2 - x1) = {x2 - x1}")
print("Slope between 2 and 3:", slope)
```

```python
x1 = 2
x2 = 3

w1 = f_non_linear(x1)
w2 = f_non_linear(x2)

slope = (w2 - w1) / (x2 - x1)

print(f"(y2 - y1) = {w2 - w1} / (x2 - x1) = {x2 - x1}")
print("Slope between 2 and 3:", slope)
```

## Numerical derivate

```python
# 1e-5 = 1 x 10 ** -5 = 0.00001
#e.g (x = 3)

def derivate_linear(x, h=1e-5):
    print('------------')

    print(f"f(x + h) = f(3 + 1 x 10 ^5) -> f(3.00001) = {f_linear(x + h)}")
    print(f"f(x) = f(3) = {f_linear(x)}")
    print(f"f(x + h) - f(x) = {f_linear(x + h) - f_linear(x)}")
    print('\noutput: ')
    return (f_linear(x + h) - f_linear(x)) / h
```

```python
print(derivate_linear(3))
print(derivate_linear(100))
```

```python
def derivate_non_linear(x, h=1e-5):
    print('------------')

    print(f"f(x + h) = f(3 + 1 x 10 ^5) -> f(3.00001) = {f_linear(x + h)}")
    print(f"f(x) = f(3) = {f_linear(x)}")
    print(f"f(x + h) - f(x) = {f_linear(x + h) - f_linear(x)}")
    print('\noutput: ')
    return (f_non_linear(x + h) - f_non_linear(x)) / h
```

```python
print(derivate_non_linear(3))
print(derivate_non_linear(4))
print(derivate_non_linear(5))
```
